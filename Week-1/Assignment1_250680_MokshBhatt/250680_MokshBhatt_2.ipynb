```python
arr = [7, 2, 9, 4, 1]

max_val = arr[0]
min_val = arr[0]

for x in arr[1:]:
    if x > max_val:
        max_val = x
    if x < min_val:
        min_val = x

print("Maximum:", max_val)
print("Minimum:", min_val)
```

```python
s = input("Enter a string: ")

left = 0
right = len(s) - 1

is_palindrome = True

while left < right:
    if s[left] != s[right]:
        is_palindrome = False
        break
    left += 1
    right -= 1

if is_palindrome:
    print("Palindrome")
else:
    print("Not a palindrome")
```

```python
arr = [5, 29, 3, 2, 50, 80]
x = 78

arr.sort()

i, j = 0, 1
n = len(arr)

while i < n and j < n:
    diff = arr[j] - arr[i]

    if diff == x:
        print("Pair found:", "(",arr[i],",", arr[j],")")
        break
    elif diff < x:
        j += 1
    else:
        i += 1
```
