#### Question 1 

```python
arr = []
n = int(input())
for i in range(n):
    temp = int(input())
    arr.append(temp)
    
maxim = arr[0]
mn = arr[0]
for i in range(n):
    if arr[i] > maxim:
        maxim = arr[i]
        
    if arr[i] < mn:
        mn = arr[i]
    

print(maxim, end = " ")
print(mn)
```

#### Question 2

```python
s = input()
isfound = True
n = len(s)
for i in range(n//2):
    if s[i] == s[n-i-1]:
        isfound = True
    else:
        isfound = False
        break

if isfound == True:
    print("YES")
else:
    print("NO")
```

```python
arr1 = []
n = int(input())

a = int(input())
b = int(input())
c = int(input())

for i in range(n):
    temp = int(input())
    arr1.append(temp)


mi = min(a, b, c)
mx = max(a, b, c)
m = (a+b+c) - (mi + mx)
l = 0
r = n-1
mid = 0
while mid <= r:
    if arr1[mid] == m:
        mid += 1
    elif arr1[mid] == mi:
        arr1[l], arr1[mid] = arr1[mid], arr1[l]
        mid+=1 
        l+=1
    elif arr1[mid] == mx:
        arr1[r], arr1[mid] = arr1[mid], arr1[r]
        r -= 1 
        
print(arr1)


    
```
